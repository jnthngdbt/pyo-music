```python
import numpy as np
import pandas as pd
import bokeh.io as bkio
import bokeh.plotting as bkpl
import bokeh.models as bkmd
```

```python
bkio.output_notebook()
```

```python
N = 200
data = pd.DataFrame({'a': np.random.randn(N), 'b':np.random.randn(N)})
```

```python
p = bkpl.figure()

c = p.circle(
    x=data.a.data,
    y=data.b.data,
    radius=0.1,
    alpha=0.5,
    hover_alpha=0.95,
    hover_color='red')

t = bkmd.HoverTool(tooltips=None, renderers=[c])

p.add_tools(t)

bkpl.show(p)
```
