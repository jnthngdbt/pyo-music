```python
import numpy as np
import pandas as pd
import bokeh.io as bkio
import bokeh.plotting as bkpl
import bokeh.models as bkmd
```

```python
bkio.output_notebook()
```

```python
Nl = 100
Nc = 10
data = pd.DataFrame(np.random.randn(Nl,Nc))
```

```python
p = bkpl.figure(width=700, height=300)

L = []
for i in np.arange(Nc):
    l = p.line(
        x=np.arange(Nl),
        y=data[i],
        line_width=2,
        alpha=0.5,
        color=(200,200,200),
        hover_alpha=0.5,
        hover_color='red')
    L.append(l)

t = bkmd.HoverTool(tooltips=None, renderers=L)
t.point_policy='snap_to_data'
t.line_policy='nearest'#'prev'

p.add_tools(t)

bkpl.show(p)

```
