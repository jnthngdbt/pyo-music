```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

import matplotlib
matplotlib.style.use('ggplot')

%matplotlib notebook
```

```python
# 100Hz 51 Woman extracted from 03:15:00 to 03:45:00 C3-A1 52 115 
eeg = pd.read_csv(os.path.join('data','eeg.txt'))
eeg[0:10].T # Show the first 10 data points
```

```python
# Get one signal vector and set the time vector.
x = eeg['C3A1']
fs = 100
N = len(x)
t = np.arange(N)/fs
x.index = t

print("That signal: %dHz, %d samples, %d secs" % (fs,  N,  np.round(t[-1])))
```

```python
def ploteeg(t,x):
    plt.figure(figsize=0.6*plt.figaspect(0.25))
    plt.plot(t,x)
    plt.xlabel('time (s)')
    plt.ylabel('uV')
    plt.ylim(-150,150)
    plt.tight_layout()
    
ploteeg(t,x)
```

```python
# Plot a few seconds snippet of the same signal
snip = (t>200) & (t<210)
ploteeg(t[snip], x[snip])
plt.axis('tight')

snip = (t>800) & (t<810)
ploteeg(t[snip], x[snip])
plt.axis('tight')
```

```python
Nw = 100                    # Number of wanted windows.
w = np.reshape(x, (Nw, -1)) # Number windows X number of time sample matrix.
Nt = w.shape[1]             # Number of time samples per window.
tw = t[:Nt]                 # Time vector for each windows.
```

```python
plt.figure(figsize=0.8*plt.figaspect(0.5))
plt.imshow(np.log(w*w), extent=(tw[0],tw[-1],t[-1],0))
plt.axis('tight')
plt.xlabel('time (s)')
plt.ylabel('windows position (s)')
plt.tight_layout()
```

```python
W = np.fft.fft(w) # By default, applied on last axe (axes=-1)
W = np.fft.fftshift(W, axes=-1) # By default, applied on ALL axes
f = np.fft.fftfreq(Nt, 1/fs)
f = np.fft.fftshift(f, axes=-1)
```

```python
# Ah, I just did a non-windowed spectrogram. I will skip negative
# frequencies and transpose it to have a freq x time representation.
i = (f >= 0)
W = W[:,i].T
f = f[i]
```

```python
plt.figure(figsize=0.8*plt.figaspect(0.5))
plt.imshow(np.log(np.abs(W)), extent=(0, t[-1], f[-1], f[0]))
plt.axis('tight')
plt.xlabel('time (s)')
plt.ylabel('frequency (Hz)')
plt.tight_layout()
```
